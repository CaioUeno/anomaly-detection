```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv('../data/brazil_covid19_macro.csv', index_col=['date']).fillna(0)
data = data.drop(['country', 'week'], axis=1)
```

```python
data.head()
```

```python
data.diff().plot(y='deaths', color='crimson', figsize=(16, 6))
plt.grid()
plt.title('Deaths over time in Brazil (diff)', fontsize=16)
plt.xlabel('Days', fontsize=14)
plt.show()
```

```python
deaths = data['deaths'].diff().fillna(0)
```

```python
deaths.to_csv('../data/deaths_data.csv')
```
